```python
try:
    meal_cost = float(input("Cost of meal: "))
    if meal_cost <= 0:
        raise ValueError("Cost must be greater than zero.")
    
    tip_percent = int(input("Tip percent: "))
    
    tip_amount = meal_cost * (tip_percent / 100)
    total = meal_cost + tip_amount

    print(f"Tip amount: {tip_amount:.2f}")
    print(f"Total amount: {total:.2f}")

except ValueError as e:
    print("Error:", e)
```

```python
FILENAME = "wizard_inventory.txt"

def display_inventory(inventory):
    print("Wizard Inventory")
    print("----------------")
    for i, item in enumerate(inventory, start=1):
        print(f"{i}. {item}")
    print()

def add_item(inventory):
    item = input("Item: ").strip()
    if item == "":
        print("Error: Item name can't be empty.\n")
    else:
        inventory.append(item)
        print(f"{item} was added.\n")

def delete_item(inventory):
    try:
        number = int(input("Number: "))
        if number < 1 or number > len(inventory):
            print("Error: Invalid item number.\n")
        else:
            removed_item = inventory.pop(number - 1)
            print(f"{removed_item} was deleted.\n")
    except ValueError:
        print("Error: Number must be a valid integer.\n")

def save_inventory(inventory):
    try:
        with open(FILENAME, "w") as file:
            for item in inventory:
                file.write(item + "\n")
    except IOError:
        print("Error: Could not save inventory file.")

def load_inventory():
    inventory = []
    try:
        with open(FILENAME, "r") as file:
            for line in file:
                inventory.append(line.strip())
    except FileNotFoundError:
        print("Warning: Inventory file not found. Starting with an empty inventory.\n")
    except IOError:
        print("Error: Problem reading inventory file.\n")
    return inventory

def main():
    inventory = load_inventory()

    while True:
        display_inventory(inventory)
        print("COMMAND MENU")
        print("show - Show all items")
        print("add - Add an item")
        print("del - Delete an item")
        print("exit - Exit program")
        print()

        command = input("Command: ").strip().lower()
        print()

        if command == "show":
            continue
        elif command == "add":
            add_item(inventory)
        elif command == "del":
            if len(inventory) == 0:
                print("Error: Inventory is empty.\n")
            else:
                delete_item(inventory)
        elif command == "exit":
            save_inventory(inventory)
            print("Bye!")
            break
        else:
            print("Error: Invalid command.\n")

main()
```
